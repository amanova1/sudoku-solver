# Sudoku Solver in Python Using Backtracking Approach

## Introduction
Sudoku is a popular logic puzzle that is played on a 9x9 grid. The goal is to fill all empty cells with numbers from 1 to 9 ensuring that each number appears only once in every row, column, and 3x3 subgrid.

I chose Sudoku for this project because it is easy to understand as a game but not so easy to solve algorithmically. At first I was not sure which topic to choose and I was considering other options such as Dobble. Sudoku seemed like a good balance between being understandable and still challenging. Also it is a good example of how mathematical rules can be translated into code.

The solution uses a backtracking approach because it is relatively easy to  implement, guarantees finding a solution and provides a base for optimizations through other techniques such as constraint propagation, Dancing Links and SAT solvers. 

## Mathematical Formulation
Sudoku can be viewed as a **constraint satisfaction problem (CSP)**. The board is represented as a $9 \times 9$ matrix where $a_{ij}=0$ indicates an empty cell.
Each cell must be filled with a value from the set $D = \{1, 2, ..., 9\}$

The number must be unique in its corresponding:
1. **row:** Each number $n$ appears exactly once in every row $i$:
$$\forall i, n: \sum_{j=1}^{9} [a_{ij} = n] = 1$$
2. **column:** Each number $n$ appears exactly once in every column $j$:
$$\forall j, n: \sum_{i=1}^{9} [a_{ij} = n] = 1$$
3. **subgrid:** Each number $n$ appears exactly once in every $3 \times 3$ subgrid:
$$\forall r, c \in \{0,1,2\}, n: \sum_{i=0}^{2} \sum_{j=0}^{2} [a_{(3r+i+1)(3c+j+1)} = n] = 1$$

Therefore solving Sudoku means finding values for all empty cells while following these constraints simultaneously.


## What is Backtracking
Backtracking is a recursive search algorithm that explores possible solutions step by step. It works by:
1. finding an empty cell
2. assigning a possible value from 1 to 9
3. checking if a value is valid
4. proceeding recursively if the value is valid
5. going back and trying another value if the assignment violates the constraints

This makes backtracking an effective method for solving Sudoku as the puzzle relies on rules that define valid and invalid choices.

## Find an Empty Cell
The algorithm first needs to find and empty cell. In the implementation below 0 means that the cell is empty.


```python
def find_empty(grid):
    for row_index in range(9):
        for col_index in range(9):
            if grid[row_index][col_index] == 0:
                return(row_index, col_index)
    return None
```

### Check if a Value is Valid
Before placing a number, we must check whether it satisfies the Sudoku constraints.

Thе function below checks whether a number can be placed in a given position by verifying the row, column and 3x3 subgrid constraints.

```python
def is_valid(grid, number, position):
    row_index = position[0]
    col_index = position[1]

    for col in range(9):
        if grid[row_index][col] == number:
            if col != col_index:
                return False

    for row in range(9):
        if grid[row][col_index] == number:
            if row != row_index:
                return False

    box_row = row_index // 3
    box_col = col_index // 3

    for row in range(box_row * 3, box_row * 3 + 3):
        for col in range(box_col * 3, box_col * 3 + 3):
            if grid[row][col] == number:
                if (row, col) != position:
                    return False

    return True
```

### Backtracking Solver
1. The solver tries numbers from 1 to 9 in each empty cell
2. If a number is valid, the algorithm proceeds to the next empty cell
3. If no valid number leads to a solution, the algorithm returns to the previous step and tests another value

```python
def solve(grid):
    empty_position = find_empty(grid)

    if empty_position is None:
        return True

    row_index = empty_position[0]
    col_index = empty_position[1]

    for number in range(1, 10):
        if is_valid(grid, number, (row_index, col_index)):
            grid[row_index][col_index] = number

            if solve(grid):
                return True

            grid[row_index][col_index] = 0

    return False
```

## Alternative Approaches for Solving Sudoku

Backtracking is not the only way to solve Sudoku. There are several alternative approaches that are more efficient but also more complex for implementation. 

**Constraint Propagation** reduces the number of possible values for each cell before the search begins. Instead of trying all possibilities it checks whether each value is still valid before placing it in a cell. For example if a row already contains 4 that value is automatically removed from all other empty cells in the row. This can be extended with more advanced techniques such as Forward Checking and Arc Consistency (AC-3). These techniques significantly reduce the number of choices the algorithm needs to explore compared to pure backtracking.

**Dancing Links** (introduced by Donald Knuth) solves Sudoku by transforming it into an Exact Cover problem. This is a mathematical problem where the goal is to select a set of rows from a matrix such that every column is covered exactly once. The key innovation is a special data structure that allows very fast addition and removal of elements making backtracking highly efficient. In practice this approach is significantly more efficient than backtracking.

**SAT Solvers** convert Sudoku into a Boolean Satisfiability Problem (SAT) where each constraint becomes a logical equation. Specialized algorithms then find values that satisfy all equations simultaneously. Modern SAT solvers are very powerful for large-scale constraint satisfaction problems.

### Complexity Comparison

| Approach | Practical Performance |
|---|---|
| Backtracking | Slow without reducing the search space |
| Constraint Propagation | Moderate improvement |
| Dancing Links | Very fast in practice |
| SAT Solvers | Highly optimized |

## Implement Constraint Propagation

I chose to implement Constraint Propagation as well because it is an optimization of the backtracking algorithm. 

Instead of trying all values from $D = \{1, 2, ..., 9\}$ for each empty cell it first computes the set of possible values for that cell and removes impossible choices. This reduces the number of recursive steps and makes the algorithm faster:

$$P_{ij} = D \setminus (R_i \cup C_j \cup S_{ij})$$

where:
- $R_i$ - values already used in row $i$
- $C_j$- values already used in column $j$
- $S_{ij}$ - values already used in the 3×3 subgrid of cell $(i, j)$
- $\setminus$ - set difference (remove these values from $D$)
- $\cup$ - union (combine all used values)

If $|P_{ij}| = 1$, the cell has only one possible value and can be filled directly without backtracking.

### Find Possible Values for a Cell

The first step is to find all possible values for a given empty cell.

```python
def possible_values(grid, row_index, col_index):

    possible_numbers = set(range(1, 10))

    for col in range(9):
        possible_numbers.discard(grid[row_index][col])

    for row in range(9):
        possible_numbers.discard(grid[row][col_index])

    subgrid_row = row_index // 3
    subgrid_col = col_index // 3

    for row in range(subgrid_row * 3, subgrid_row * 3 + 3):
        for col in range(subgrid_col * 3, subgrid_col * 3 + 3):
            possible_numbers.discard(grid[row][col])

    return possible_numbers
```

### Fill Cells with a Single Possible Value
If a cell has only one possible value we can fill it directly without backtracking. We repeat this process until no more cells can be filled.
Note: For this test I use a simpler puzzle where Constraint Propagation alone is sufficient to find the solution.

```python
easy_grid = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

def fill_cells(grid):

    updated = True

    while updated:

        updated = False

        for row_index in range(9):
            for col_index in range(9):

                if grid[row_index][col_index] == 0:

                    numbers = possible_values(grid, row_index, col_index)

                    if len(numbers) == 1:
                        grid[row_index][col_index] = numbers.pop()
                        updated = True

    return grid

print("Sudoku grid before solving:")
print_grid(easy_grid)

print("\nSolving the Sudoku...\n")
fill_cells(easy_grid)

print("\nSolved Sudoku grid:")
print_grid(easy_grid)
```

This shows that for simple puzzles the mathematical constraints are strong enough to find the solution directly.

## Solving Sudoku Like a Human
So far we have used backtracking and constraint propagation to solve Sudoku algorithmically. However humans solve Sudoku very differently as they use logical reasoning instead of trial and error.

Backtracking is essentially trial and error. A computer can afford to try thousands of combinations per second but a human sitting with a pen and paper works completely differently. People do not guess and backtrack, they reason.

They look at a row and think: "This cell can only be a 7 because all other numbers are already taken." Or they look at a subgrid and think: "The number 3 can only fit in this one cell so it must go there."

This kind of reasoning is not just intuitive, it is mathematically precise. Each logical step is based on the same constraints we defined earlier: every number must appear exactly once in every row, column and 3×3 subgrid. This is fundamentally different from backtracking because there is no guessing but only reasoning.

In this section we implement four of the most fundamental human solving techniques using the Easy puzzle from our earlier tests. Each technique is more powerful than the previous one and together they can solve many puzzles without any backtracking at all.

1. **Naked Single**: if a cell has only one possible value fill it in directly
2. **Hidden Single**: if a number can only go in one cell within a row, column or subgrid place it there
3. **Naked Pair**: if two cells in a row share the same two possible values eliminate those values from all other cells in the row
4. **Pointing Pairs**: if a number in a subgrid can only appear in one row eliminate it from the rest of that row

### Naked Single

A Naked Single is the simplest human solving technique. If a cell has only one possible value left after eliminating all values already used in its row, column and subgrid that value must go there.

Mathematically this is the same formula we used for Constraint Propagation:

$$P_{ij} = D \setminus (R_i \cup C_j \cup S_{ij})$$

When $|P_{ij}| = 1$ there is only one option left and no guessing is needed.

**Example:** In the Easy puzzle, cell $(4, 4)$:
- Row 4 already contains: $\{8, 1, 3, 4\}$
- Column 4 already contains: $\{1, 2, 6, 7, 8, 9\}$
- The subgrid already contains: $\{8, 2, 3, 6\}$

After removing all of these from $\{1, ..., 9\}$, only one value remains:

$$P_{44} = \{5\}$$

So cell $(4, 4)$ must be **5**. 

We already implemented this in our "fill cells" function. Every time a cell has only one possible value we fill it in directly.

### Hidden Single

Naked Single looks at a cell and asks: "What values can go here?" Hidden Single works the other way around. It looks at a number and asks: "Where this value can go in this row, column or subgrid?"

Imagine you are solving the Easy puzzle and you need to place the number **4** in row 0. You check each empty cell and find that only one of them allows 4 because of the column and subgrid constraints. Even if that cell has other possible values it must be 4. There is simply nowhere else for it to go.

Mathematically for each number $n$ and each row $i$ we find all empty cells where $n$ is a valid placement:

$$H_i(n) = \{j : a_{ij} = 0 \text{ and } n \in P_{ij}\}$$

If $|H_i(n)| = 1$, then $n$ must go in that cell - even if $|P_{ij}| > 1$.

The same logic applies to columns and subgrids.

### Implemenytation of Hidden Single
We will implement the Hidden Single approach using the Easy puzzle from our earlier tests.

```python
def hidden_single(grid):

    updated = True

    while updated:
        updated = False

        # Check rows
        for row_index in range(9):
            for number in range(1, 10):

                possible_positions = []

                for col_index in range(9):
                    if grid[row_index][col_index] == 0:
                        if number in possible_values(grid, row_index, col_index):
                            possible_positions.append(col_index)
                if len(possible_positions) == 1:
                    grid[row_index][possible_positions[0]] = number
                    updated = True

        # Check columns
        for col_index in range(9):
            for number in range(1, 10):

                possible_positions = []

                for row_index in range(9):
                    if grid[row_index][col_index] == 0:
                        if number in possible_values(grid, row_index, col_index):
                            possible_positions.append(row_index)
                if len(possible_positions) == 1:
                    grid[possible_positions[0]][col_index] = number
                    updated = True

        # Check subgrids
        for subgrid_row in range(3):
            for subgrid_col in range(3):
                for number in range(1, 10):

                    possible_positions = []

                    for row_index in range(subgrid_row * 3, subgrid_row * 3 + 3):
                        for col_index in range(subgrid_col * 3, subgrid_col * 3 + 3):
                            if grid[row_index][col_index] == 0:
                                if number in possible_values(grid, row_index, col_index):
                                    possible_positions.append((row_index, col_index))
                    if len(possible_positions) == 1:
                        row_index = possible_positions[0][0]
                        col_index = possible_positions[0][1]
                        grid[row_index][col_index] = number
                        updated = True

    return grid
```

```python
easy_grid = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],
    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],
    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

print("Sudoku grid before solving:")
print_grid(easy_grid)

print("\nSolving the Sudoku...\n")

hidden_single(easy_grid)

print("Solved Sudoku grid:")
print_grid(easy_grid)
```

### Observation

Hidden Single is enough to solve the Easy puzzle completely. This shows that for simpler puzzles human logical reasoning is not only intuitive but also mathematically sufficient and no backtracking is needed.

Compared to Naked Single Hidden Single is more powerful because it looks at the puzzle from a different angle. Instead of asking what values a cell can have it asks where a value can go. Together these two techniques can solve a wide range of puzzles without any trial and error.

### Naked Pair

A Naked Pair occurs when two cells in the same row column or subgrid have exactly the same two possible values. Even though we do not know which cell gets which value we know that these two values are "reserved" for these two cells. This means we can safely exclude them from all other cells in the same row column or subgrid.

**Example:** If two cells in a row both have possible values $\{3, 7\}$. One of them will be 3 and the other will be 7 and we just don't know which is which yet. But we know that no other cell in that row can be 3 or 7.

Mathematically for two cells $(i, j_1)$ and $(i, j_2)$ in row $i$:

$$P_{ij_1} = P_{ij_2} = \{a, b\}$$

Then for all other empty cells $(i, j)$ where $j \neq j_1$ and $j \neq j_2$:

$$P_{ij} = P_{ij} \setminus \{a, b\}$$

The same logic applies to columns and subgrids.

### Pointing Pairs

A Pointing Pair occurs when a number can only appear in one row within a subgrid. Since that number must go somewhere in that subgrid and all its possible positions are in the same row we can exclude it from the rest of that row outside the subgrid.

**Example:** If the number 5 can only go in cells from row 1 within the top left subgrid then 5 cannot appear anywhere else in row 1.

Mathematically for a number $n$ and a subgrid $S_{rc}$, let $R(n, S_{rc})$ be the set of rows where $n$ is possible within the subgrid:

$$R(n, S_{rc}) = \{i : \exists j \text{ such that } a_{ij} = 0 \text{ and } n \in P_{ij}, (i,j) \in S_{rc}\}$$

If $|R(n, S_{rc})| = 1$, then $n$ can be excluded from all cells in that row outside the subgrid.

## Complexity and Analysis

The backtracking algorithm is simple and effective but it may become slow for difficult Sudoku puzzles because it explores many possible combinations.

In the worst case the time complexity is very high because the algorithm may need to try many values recursively. However for standard Sudoku puzzles backtracking is often fast enough and provides a clear example of recursive problem solving.

The results from our tests show an interesting observation: the difficulty of a puzzle does not depend only on the number of given cells, but also on their distribution. In our tests, the Medium puzzle had fewer backtracking steps (37) than the Easy puzzle (151) even though it had fewer given cells (16 vs 32). This suggests that the arrangement of the  known cells plays a key role in how efficiently the algorithm finds a solution.
