# Sudoku Solver in Python Using Backtracking Approach

## Introduction
Sudoku is a popular logic puzzle that is played on a 9x9 grid. The goal is to fill all empty cells with numbers from 1 to 9 ensuring that each number appears only once in every row, column, and 3x3 subgrid.

I chose Sudoku for this project because it is easy to understand as a game but not so easy to solve algorithmically. At first I was not sure which topic to choose and I was considering other options such as Dobble. Sudoku seemed like a good balance between being understandable and still challenging. Also it is a good example of how mathematical rules can be translated into code.

The solution uses a backtracking approach because it is relatively easy to  implement, guarantees finding a solution and provides a base for optimizations through other techniques such as constraint propagation, Dancing Links and SAT solvers. 

## Mathematical Formulation
Sudoku can be viewed as a constraint satisfaction problem. Each empty cell must satisfy the following constraints.

The number must be unique in its corresponding:
1. row
2. column
3. 3x3 subgrid

Therefore solving Sudoku means finding values for all empty cells while following these constrains.
The Sudoku board can be represented as a 9x9 matrix where 0 means that the cell is empty.

## What is Backtracking
Backtracking is a recursive search algorithm that explores possible solutions step by step. It works by:
1. finding an empty cell
2. assigning a possible value from 1 to 9
3. checking if the value is valid
4. proceeding recursively if the value is valid
5. going back and trying another value if the assignment violates the constraints

This makes backtracking an effective method for solving Sudoku as the puzzle relies on rules that define valid and invalid choices.

## Example Sudoku Grid

This Sudoku example represents a medium to hard difficulty level, as it contains few initial values and requires deeper recursive exploration to reach a solution.

```python
grid = [
    [0, 0, 0, 0, 0, 0, 2, 0, 0],
    [0, 8, 0, 0, 0, 7, 0, 9, 0],
    [6, 0, 2, 0, 0, 0, 5, 0, 0],
    [0, 7, 0, 0, 6, 0, 0, 0, 0],
    [0, 0, 0, 9, 0, 1, 0, 0, 0],
    [0, 0, 0, 0, 2, 0, 0, 4, 0],
    [0, 0, 5, 0, 0, 0, 6, 0, 3],
    [0, 9, 0, 4, 0, 0, 0, 7, 0],
    [0, 0, 6, 0, 0, 0, 0, 0, 0]
]
```

```python
def print_grid(grid):
    for row_index in range(len(grid)):
        if row_index % 3 == 0 and row_index != 0:
            print("-" * 21)
        
        for col_index in range(len(grid[0])):
            if col_index % 3 == 0 and col_index != 0:
                print("|", end=" ")
            
            if col_index == 8:
                print(grid[row_index][col_index])
            else:
                print(grid[row_index][col_index], end=" ")

print("Example Sudoku grid before solving:")
print_grid(grid)
```
